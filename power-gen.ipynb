```python
import requests
import pandas as pd
```

```python
# assign user agent to variable in case it's needed

my_ua = {"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:94.0) Gecko/20100101 Firefox/94.0"}
```

```python
from datetime import datetime, timedelta

today = datetime.today().strftime('%Y-%m-%d')
yesterday = (datetime.today() - timedelta( days=1 )).strftime('%Y-%m-%d')
dayb4yesterday = (datetime.today() - timedelta(days = 2)).strftime('%Y-%m-%d')

today, yesterday, dayb4yesterday
```

```python
url = 'https://data.elexon.co.uk/bmrs/api/v1/generation/outturn/current?fuelType=COAL&fuelType=CCGT&fuelType=COAL&fuelType=INTELE&fuelType=INTELEC&fuelType=INTEW&fuelType=INTFR&fuelType=INTGRNL&fuelType=INTIFA2&fuelType=INTIRL&fuelType=INTNED&fuelType=INTNEM&fuelType=INTNSL&fuelType=INTVKL&fuelType=NPSHYD&fuelType=NUCLEAR&fuelType=OCGT&fuelType=OIL&fuelType=OTHER&fuelType=PS&fuelType=WIND&format=json'

req = requests.get(url, headers = my_ua)

req.status_code
```

```python
req.json()
```

```python
big_list = []

for i in req.json():
    small_list = []
    for x,y in i.items():
        if (x != 'dataset'):
            small_list.append(y)
    big_list.append(small_list)
big_list
```

```python
cols = ['fuelType', 'currentUsage', 'currentPercentage', 'halfHourUsage', 'halfHourPercentage', 'twentyFourHourUsage', 'twentyFourHourPercentage']

df = pd.DataFrame(big_list, columns = cols)

df
```
